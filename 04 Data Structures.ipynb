```python
metals = ['silver','copper','gold','aluninium','gold','silver',
         'gold','copper','iron','gold','copper']
```

```python
# Create a new list using above list which will contain same items in the
# same order except the item 'copper'
```

```python
new_metals = []
for metal in metals:
    if metal != 'copper':
        new_metals.append(metal)
```

```python
new_metals
```

```python
new_metals = [metal for metal in metals   if metal != 'copper']
```

```python
new_metals
```

```python
# Read the information of students like- roll no, name and marks
# Find the name of topper.
```

```python
n = int(input('How many students? '))
studs = []
for i in range(n):
    roll = int(input('Enter roll number:'))
    name = input('Enter name:')
    marks = float(input('Enter marks:'))
    studs.append((roll, name, marks))
```

```python
studs
```

```python
m = studs[0][2]
m
```

```python
for s in studs:
    if s[2] > m:
        m = s[2]
        topper = s[1]
```
