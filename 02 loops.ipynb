# while

```python
count = 0

while count < 5:
    print("krunal")
    count += 1
```

```python
# find the addition of numers from 1 to 10
```

```python
count = 1
add = 0

while count <= 10:
    add = add + count
    count += 1

print("addition is", add)
```

# loop control statements

```python
for i in range(1, 11):
    print(i)
    if i == 6:
        break
```

```python
for i in range(1, 11):
    
    if i == 6:
        break
    print(i)
```

```python
# continue
```

```python
for i in range(1, 11):
    
    if i == 6:
        continue
    print(i)
```
