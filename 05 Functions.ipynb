#### Parameterized Functions

```python
def square(n):
    sq = n *n
    print("Square is:", sq)
```

```python
def square(n):
    sq = n *n
    return(sq)
```

```python
square(67)
```

```python
5 * square(5)
```

#### Multi-return statement

```python
def calc(num):
    sq = num ** 2
    cu = num ** 3
    return sq, cu
```

```python
calc(6)
```

```python
# Read a number from user. write and call a function to find 
# factorial of it
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact = fact * i 
    return fact
```

```python
n = int(input("Enter the number:")) 
factorial(n)
```
