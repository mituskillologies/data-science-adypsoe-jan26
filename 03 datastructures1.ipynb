```python
l = [12, 55, 90, 67, 55, 12, 13]
```

```python
"""
iterate over the list and print their index and element
eg...
0 -> 12
1 -> 55
2 -> 90
and so on...
""";
```

```python
for i in l:
    print(f"{l.index(i)} -> {i}")
```

```python
list(enumerate(l))
```

```python
for ind, val in enumerate(l):
    print(f"{ind} -> {val}")
```

```python
l
```

```python
for i in l:
    print(i * i)
```

```python
for i in l:
    if i % 2 == 0:
        print(i)
```
