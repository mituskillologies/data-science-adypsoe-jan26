```python
class Person:                         # class definition
    def __init__(self, n, a):         # constructor
        self.name = n                 # instance variables
        self.age = a

    def show(self):                   # class function / method
        print("Values are:", self.name, self.age)
```

```python
# create the object
p1 = Person("Ajay", 23)
```

```python
p1.show()
```

```python
p2 = Person("Priya", 25)
p2.show()
```
