```python
# time difference for list and array
```

```python
l1 = list(range(1, 1000000))
a = np.array(l1)
```

```python
# list time
start_time = time.time()
newl = [x*2 for x in l1]
end_time = time.time()
time_list = end_time - start_time
print("time taken by list:", time_list)
```

```python
# array time
start_time = time.time()
newa = a * 2
end_time = time.time()
time_array = end_time - start_time
print("time taken by array:", time_array)
```
